```python
from glob import glob
import csv
import os
import shutil
```

```python
root_dir = "R:/AIMIRA/AIMIRA_Database/LUMC/"
all_dir = os.listdir(root_dir) 
all_dir = [i for i in all_dir if 'TRT' in i]
```

```python
all_dir
```

```python
file_fpath_ls = []
for folder in all_dir:
    dt =  {'id': folder}
    first_visit = os.listdir(root_dir+folder) [0]
    view_names = ['COR', 'TRA']
    for view in view_names:
        three_files = glob(root_dir + folder + '/' + first_visit + f'/*Post{view}*/images/itk/*.mha')
        print(three_files)
        for i in three_files:
            # if 'Foot' in i:
            #     dt.update({f'MTP_{view}': i})
            # if 'Wrist' in i:
            #     dt.update({f'WRI_{view}': i})
            if 'MCP' in i:
                dt.update({f'MCP_{view}': i})
            # else:
            #     raise Exception(f'wrong file name: {i}. File name should include at least Foot, Wrist or MCP')

    file_fpath_ls.append(dt)
    
```

```python
dest_dir = "E:/jjia/data/aimira/TRT/"
for file_dt in file_fpath_ls:
    for k, v in file_dt.items():
        if k != 'id':
            source_file = v
            destination_file = dest_dir + v.split("\\")[-1]
            print(destination_file)
            shutil.copy(source_file, destination_file)
```
